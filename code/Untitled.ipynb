```python
def date_features(train):
    train['release_date'] = pd.to_datetime(train['release_date'])  # normaliztion date
    train['release_year'] = train['release_date'].dt.year  # extract year
    train['release_month'] = train['release_date'].dt.month     # extract month
    train['release_day'] = train['release_date'].dt.day  # extract dat
    train['release_quarter'] = train['release_date'].dt.quarter  # extract season
    return train


train = date_features(train)
train['release_year'].head()
```

```python
#deal with the outliers
train['release_year']=np.where(
    train['release_year']>2019, train['release_year']-100,train['release_year'])
train['release_year'].iloc[np.where(train['release_year']>2019)][:10]
```

```python
fig = plt.figure(figsize=(10, 6))
data = dict(frequency_genres)
names = list(data.keys())
values = list(data.values())

plt.barh(sorted(range(len(data)), reverse=True),
         values, tick_label=names, color='green')
plt.xlabel('Count')
plt.title('Count of movie genre')
plt.show()
```

```python
list_of_cast_names = list(train1['cast'].apply(
    lambda x: [i['name'] for i in eval(x)] if type(x) == str else []).values)
most_common_keywords = Counter(
    [i for j in list_of_cast_names for i in j]).most_common(20)

fig = plt.figure(figsize=(10, 6))
data = dict(most_common_keywords)
names = list(data.keys())
values = list(data.values())

plt.barh(sorted(range(len(data)), reverse=True),
         values, tick_label=names, color='grey')
plt.xlabel('Count')
plt.title('Top 20 polular actors')
plt.show()
```

```python
train1['num_cast'] = train1['cast'].apply(
    lambda x: len(eval(x)) if type(x) == str else 0)
train1['all_cast'] = train1['cast'].apply(lambda x: ' '.join(
    sorted([i['name'] for i in eval(x)])) if type(x) == str else '')
top_cast_names = [m[0] for m in Counter(
    [i for j in list_of_cast_names for i in j]).most_common(30)]
for g in top_cast_names:
    train1['cast_name_' + g] = train1['all_cast'].apply(lambda x: 1 if g in x else 0)
cols = [i for i in train1.columns if 'cast_name' in str(i)]
train1[cols].head()
```

```python
cast_name_Samuel_L_Jackson = train1.loc[train1['cast_name_Samuel L. Jackson'] == 1, ]
cast_name_Morgan_Freeman = train1.loc[train1['cast_name_Morgan Freeman'] == 1, ]
cast_name_Liam_Neeson = train1.loc[train1['cast_name_Liam Neeson'] == 1, ]
cast_name_Owen_Wilson = train1.loc[train1['cast_name_Owen Wilson'] == 1, ]

cast_name_Samuel_L_Jackson_revenue = cast_name_Samuel_L_Jackson.mean()[
    'revenue']
cast_name_Morgan_Freeman_revenue = cast_name_Morgan_Freeman.mean()['revenue']
cast_name_Liam_Neeson_revenue = cast_name_Liam_Neeson.mean()['revenue']
cast_name_Owen_Wilson_revenue = cast_name_Owen_Wilson.mean()['revenue']


cast_revenue_concat = pd.Series([cast_name_Samuel_L_Jackson_revenue,
                                 cast_name_Morgan_Freeman_revenue,
                                 cast_name_Liam_Neeson_revenue,
                                 cast_name_Owen_Wilson_revenue])

cast_revenue_concat.index = ['Samuel L. Jackson',
                             'Morgan Freeman',
                             'Liam Neeson',
                             'Owen_Wilson']

fig = plt.figure(figsize=(13, 5))
cast_revenue_concat.sort_values(ascending=True).plot(
    kind='barh', title='Mean Revenue (100 million dollars) by Top 4 Most Common Cast')
plt.xlabel('Revenue (100 million dollars)')
```
